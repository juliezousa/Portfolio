# Support Vector Machine
### General idea:

The goal of SVMs is to find the line that maximizes the minimum distance to the line.

### Optimal margin classifier:

The optimal margin classifier $h$ is such that:

$$
h(x) = sign(w^Tx - b),
$$

where $(w,b) \in \mathbf{R}^n\times\mathbf{R}$ is the solution of the following optimization problem:

$\min\frac{1}{2}||w^2||$ such that $y_i(w^Tx_i - b) \geq 1$.

The decision boundary is defined as $w^Tx - b = 0$.

### Hinge Loss:

The hinge loss is used in the setting of SCMs and is defined as follows:

$$
L(z,y) = [1 - yz]_+ = \max(0,1-yz)
$$

### Kernel

Given a feature mapping $\phi$, we define the kernel $K$ as follows:

K(x,z) = \phi(x)^T\phi(z).

In practice, the kernel $K$ defined by $K(x,z) = \exp\left(-\frac{||x-z||^2}{2\sigma^2}\right)$ is called the Gaussian kernal and is commonly used.


Rermark: we say that we use the "Kernel trick" to compute the cost function using the kernel because we actually don't need to know the explicit mapping $\phi$, which is often very complicated. Instead, only the values $K(x,z)$ are needed.

### Lagragian:

We define the Legragian $L(w,b)$ as follows:

$$
L(w,b) = f(w) + \sum_{i=1}^l\beta_ih_i(w)
$$.

Remark: the coefficients $\beta_i$ are called the lagrange multipliers. Often denoted by $\lambda$.

## SVM - Social network data set

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns



dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)


# Fitting

classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)

#prediction

y_pred = classifier.predict(X_test)

#confusion matrix

cm = confusion_matrix(y_test, y_pred)
```

```python
# Plot confusion matrix with labels
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=False, fmt="d", cmap="Purples", cbar=False)

# Adding labels
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Adding TP, TN, FP, FN labels
for i in range(len(cm)):
    for j in range(len(cm)):
        text = f"{cm[i, j]}"
        if i == j:
            text += f"\n\nCorrectly\nClassified"
        else:
            text += f"\n\nIncorrectly\nClassified"
        plt.text(j + 0.5, i + 0.5, text, ha='center', va='center', color='red' if i != j else 'white')

plt.show()
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('blue', 'violet')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('blue', 'violet'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('darkblue', 'purple')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('darkblue', 'purple'))(i), label = j)
plt.title('SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
